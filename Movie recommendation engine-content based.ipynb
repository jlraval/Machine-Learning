```python
import ast

def convert(df, col):
    main_list=[]
    for i in range(len(df)):
        list1=[]
        for j in ast.literal_eval(df.loc[i,col]):
            list1.append(j['name'])
        main_list.append(list1)
    return main_list
```

```python
list_of_genres=convert(movies_final, 'genres')
```

```python
movies_final['genres_new']=list_of_genres
```

```python
def convert1(df, col):
    main_list=[]
    for i in range(len(df)):
        list1=[]
        for j in ast.literal_eval(df.loc[i,col])[:3]:
            list1.append(j['name'])
        main_list.append(list1)
    return main_list
```

```python
def convert2(df, col):
    main_list=[]
    for i in range(len(df)):
        list1=[]
        for j in ast.literal_eval(df.loc[i,col]):
            if j['job'] == 'Director':
                list1.append(j['name'])
                break
        main_list.append(list1)
    return main_list
```

```python
# the last thing left to do is to join the name and surname into one word, otherwise in BOW they will be treated as two separate words and that would lead to incorrect vector space generated. it would be meaningless to have two independent vectors formed by name and surname individually as they individually do not convery meaning. also, with two diff names in the data, the model will be confused. so it is best to have name-surname together to have a unique unit vector.
# we will join  2 words to 1 word in all the cols
```

```python
movies_final['genres_new']=movies_final['genres_new'].apply(lambda x: [i.replace(" ", "") for i in x])  # apply needs a function. so lambda must be there. and then we wanted the output as a list, so we used list comprehension as the stmt of the lambda function.
```

```python
movies_final['keywords_new']=movies_final['keywords_new'].apply(lambda x: [i.replace(" ", "") for i in x])  # apply needs a function. so lambda must be there. and then we wanted the output as a list, so we used list comprehension as the stmt of the lambda function.
```

```python
movies_final['cast_new']=movies_final['cast_new'].apply(lambda x: [i.replace(" ", "") for i in x])  # apply needs a function. so lambda must be there. and then we wanted the output as a list, so we used list comprehension as the stmt of the lambda function.
```

```python
movies_final['crew_new']=movies_final['crew_new'].apply(lambda x: [i.replace(" ", "") for i in x])  # apply needs a function. so lambda must be there. and then we wanted the output as a list, so we used list comprehension as the stmt of the lambda function.
```

```python
# now our challenge is to identify the similarity score between two tags to identify the similar movies.
# i.e. we need to identify the similarity between the texts
```

```python
# now to find similarity between two sentences --> 1) one approach is to find common words between the sentences. if the num of common words are large, then the sentences are similar to each other.
# but this is not a very good approach as it just simply counts the number of common words in two sentences. but BOW is better as it also takes into account the frequency of each word appearing in the two sentences to identify the similarity. And this is done using vectorization. BOW is one of the methods for vectorization : i.e.converting a text to a vector. 
```

```python
# and we will remove the stop words from our text. this is internally done inside countvectorizer. BOW is implemented using countvectorizer.
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english') 
cv.fit_transform(new_df['tags'])
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray() # by default it outputs a compressed sparse matrix. and to view the output matrix we need to convert it to array. 
```

```python
# now using stemming on the original text data as preprocessing and then apllying BOW. stemming is important as it reduces redundancy of words. eg., loved, loving, love will all be converted to love after stemming. so now only one feature love which is in reality the real feature for all three words and not taking loved, loving and love as 3 diff features. i.e. finding the root word 
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
stem(new_df['tags'][1]) # so stem works well for a given row data
```

```python
# applying the stemming code to the whole tags col noq
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
# again doing BOW using count vectorizer on this
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english') 
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray() # by default it outputs a compressed sparse matrix. and to view the output matrix we need to convert it to array. 
```

```python
vectors
```

```python
# applying similarity (cosine similarity)
```

```python
# now we need to calculate the distance of one movie with all the other movies. the smaller the distance, the more is the similarity between those two movies.
# but note, that we are not calculating the euclidean distance ebtween them (as like in clustering), instead we are calculating the cosine distance i.e., the cosine angle between the vectors. the smaller the angle, the more similar they are.
# the reason of using cosine angle and not clustering is because in higher dimensional data, euclidean distance actually fails. that is why we cannot identify vectors close to each other based on euclidean distance. and that is why we use cosine angle. knn also fails in higher dimension simply because it uses euclidean distance. clustering fails in higher dimension coz of euclidean dist and hence clustering cannot be used for recomendation purpose.
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
result = cosine_similarity(vectors) # so for each vector, they will find the dist of that vector with all the other vectors.
```

```python
result.shape
```

```python
result
```

```python
# recommender function
```

```python
def recommend(inp): # input to the function is a movie name
    movie_index = (np.where(new_df['title']==inp))[0][0]
   # index = new_df[new_df['title']=='Avatar'].index[0] 
    top_five_movie_index=(np.argsort(result[movie_index]))[-1:-7:-1] # argsort sorts with largest dist index at the top. hence i had to take the last 6 indexes except the last one.
    for i in top_five_movie_index:
        print(new_df['title'][i])
```
